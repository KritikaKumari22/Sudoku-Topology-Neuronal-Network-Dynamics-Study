```python
from brian2 import *
%matplotlib inline
from math import *
start_scope()
N = 10                          # No. of neurons
tau = 10*ms                     # Time constant of neuron
A_osc = 0.2                     # Oscillation amplitude
freq = 25*Hz                    # Frequency of common oscillatory drive
#transm_delay = pow(10,-5)*ms    # Delay between firing and reception of pulse by post-synaptic neuron or the Refractory period
#Ib = 1
defaultclock.dt = 0.5*ms
#C_inh = 0
#C_exc = 0
P = PoissonGroup(1, rates=10*Hz)
eqs = '''
dv/dt = v/tau : 1               
I : 1
'''
# Create a neuron group 'G' following the above equation
G = NeuronGroup(N, eqs, threshold='v>0', reset='v=0', method='euler')

#G = NeuronGroup(N, eqs, 'rates : Hz', threshold='rand()<rates*dt')
S = Synapses(P, G, on_pre='v += 0.1')
S.connect()

#background_rateE = 2*2000*Hz
#B_E = PoissonGroup(N, background_rateE)

G.v = 0           #Initial values of voltage for the neurons randomly provided across neurons

# Record of potential for all neurons
potential = StateMonitor(G, 'v', record=True)

run(5000*ms)
```

```python
figure(figsize=(15,2))
title('Neuron 1 Experiencing Poisson Input 0.1 mV', fontsize=16)
plot(potential.t/ms, potential.v[0])
xlabel('Time (ms)', fontsize=16)
ylabel('Potential (mV)', fontsize=16)
show()
```
