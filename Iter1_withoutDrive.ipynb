### _Preface_ <br>
<br>
Neurons, through excitatory and inhibitory connections with other neurons form a network which can generate a spatio-temporal dynamics of specific sub-set of neurons fire together(or separately) at a given point in time by virtue of the stimulus they have received and communicated to their specific connections which provide the localization in space to these firings. Inhibitory synaptic connections prevent the post-synaptic neurons from firing by decreasing it's membrane potential, whereas excitatory connections do the opposite; in the real nervous system a neuron can synapse with multiple neurons who can provide either of such inhibitory/excitatory signals which can have their own post-synaptic neuron voltage change weights and thus, generation of action potential will depend on the net stimulus received from all the pre-synaptic neurons. In a purely inhibitory network, stoppage of a group of neurons spiking in synchrony will release others from inhibition that will fire till the point when another group with inhibitory connection with them start to fire. The identity of synchronously spiking neurons and the order in which they fire are determined by the structure of the network [1]. 

Such a biological design of neurons and nerve impulse conduction allows modelling of this phenomena as networks or graphs, e.g.- each neuron depicting a vertex and the connections formed between them is denoted by an edge; in such a Graphical representation of the network neurons that spike together can be depicted by one coloring of the vertices representing them. Thus, spatiotemporal patterns formed by spiking are imitated as colourings of certain vertices in time. In such a model, the read-out or the output should also correspond to biologically relevant perception that is generated with respect to the stimuli received; and therefore for a specific perception e.g.-a specific odor perception by the olfactory neurons, should correspond to a specific spatio-temporal pattern generated in the brain. Therefore, a specific sequence of spiking in sub-sets is called a 'Periodic attractor' and by the property of it's network is the result of the initial-stimulus and the excitatory/inhibitory weights in the synapses and the configuration made by the number of synapses to and from every neuron.

### The Model/ Methodology 

My project aims at working with a network which can also be modelled in the form of a graph with the following properties: -
* Neurons as vertices of the graph
* Edges representing reciprocal(both ways)-coupling 
* Excitatory and Inhibitory edge sets form complementary graphs of each other
* The neuorons modeled as Mirollo-Strogatz pulse-coupled oscillators [3], a generalization of the integrate and fire neuron [4], where time-evolution of the voltage is given by the equation
$$ \frac{dv_i}{dt} = I  v_i , 0  v_i  1, i = 1, . . . , N $$
<br>
<center>  Units v_i: Voltage, where  is constant and I is the Driving Current or the Input Stimulus  </center>
<br>
<center> The Voltage evolved with the equation: $ x_i = \frac{I}{}(1-e^{t}) ,  0 < t  T $ </center>
<br>
*  When $v_i$ reached threshold, = 1, it emitted a pulse and was reset to 0.

* Phase _()_ of the of the voltage, is denoted by $ F() = v\frac{T}{}$

* A pulsatile interaction between oscillators or the Synaptic Coupling is modelled as shift in the Phase.

* The shift in the phase of the post-synaptic neuron by incoming pulses is given by the Heaviside Step Function: -
$$ _j (t^+) = min"\{"U1(U(_j (t)) + \epsilon_e), 1"\}"  for \, excitatory \, pulse(s) \, \epsilon_e $$
$$ _j (t^+) = min"\{"U1(U(_j (t)) - \epsilon_e), 0"\}"  for \, inhibitory \, pulse(s) \, \epsilon_e $$

* An oscillatory Driving Current, which is a simple sinusoidal-wave is also added along with the inut stimulus at the same time-steps. This could be thought of as physiological noise while receiving the Odor stimulus from the environment as it comes in bursts but the chemical conduction of the odor into concentration gradient change leading to input stimulus generation could be easier or harder depending on the number of receptors are free.

```python
from brian2 import *
%matplotlib inline
import numpy as np
from numpy import heaviside
from math import *
import pandas as pd
#from numba import jit

#@jit
# Potential function
def U(phi,I=1.01,gamma=1,T=4.615120516841259, tau=1):
    value = (I/gamma)*(1-np.exp(-gamma*T*phi/tau))
    return value
#@jit
# Function to convert potential to 'phase' corresponding to Mirollo-Strogatz model
def U_inv(y,I=1.01,gamma=1,T=4.615120516841259,tau=1):
    value = (tau/(gamma*T))*np.log((1-(gamma*y/I))**(-1))
    return value
```

```python
## Loading Inhibitory and Excitatory Edge topology from a 81  81 Matrix specifying 20 connections for each neuron 
# Import Sudoku matrices: (rows: pre-synaptic, columns: post-synaptic)
inhib_connect = np.loadtxt('Clueless_Sudoku_inhib.txt', dtype=int)
excit_connect = np.loadtxt('Clueless_Sudoku_excit.txt', dtype=int)

# Get tuples of corresponding connection indices for putting in Brian2's synapse connect() function
inhib_pre, inhib_post = inhib_connect.nonzero()
excit_pre, excit_post = excit_connect.nonzero()
```

```python
start_scope()
N = 81                          # No. of neurons
tau = 10*ms                     # Time constant of neuron
#A_osc = 0.2                     # Oscillation amplitude
#freq = 25*Hz                    # Frequency of common oscillatory drive
transm_delay = pow(10,-5)*ms    # Delay between firing and reception of pulse by post-synaptic neuron or the Refractory period
 
# Pulse coupling strengths: inhibitory and excitatory
C_inh = -2.7*pow(10,-2)
C_exc = +2.8*pow(10,-3)

eqs = '''
dv/dt = I+(1-v)/tau : 1               
I : 1
'''

# Create a neuron group 'G' following the above equation
G = NeuronGroup(N, eqs, threshold='v>1', reset='v=0', method='euler')
G.v = 'rand()'            #Initial values of voltage for the neurons randomly provided across neurons
G.I = 0.1

# Inhibitory synapses
S_inhib = Synapses(G,G, on_pre='v_post += C_inh', delay=transm_delay)
S_inhib.connect(i=inhib_pre, j=inhib_post)

# Excitatory synapses
S_excit = Synapses(G,G, on_pre='v_post += C_exc', delay=transm_delay)
S_excit.connect(i=excit_pre, j=excit_post)

# Record of all spikes
firing_time = SpikeMonitor(G)
# Record of potential for all neurons
potential = StateMonitor(G, 'v', record=True)
# Record of oscillatory drive (common to all neurons)
osc = StateMonitor(G, "I", record=0)
```

```python
run(4000*ms) #where x is the time in ms for which we want to simulate the system
```

### Conclusion & Future Prospects

Neuronal networks are able to generate stable responses despite highly noisy environmental input. This project will explore invariants in the dynamics of a recurrent excitatory inhibitory network driven by noisy intermittent input.
<br>

* Building the code for an input stimulus which is a step function with a freqquency taken from a poisson distribution.
* Visualizing the phases of the spiking pattern generated with time across neurons. These are also referred to as 'solutions'.
* The solution/'Periodic Attractor'/pattern arrived at towards the end of each simulation run will be checked for whether it is a solution to Sudoku, i.e., is a minimal colouring network and when the neurons in the $9  9$ matrix are replaced by the colours they posses solve the Sudoku puzzle which will be when the colour possesed is unique in the row, column and the sub-grid of the grid-cell position occupied by the neuron. 
* To analyze the minimal-colouring solutions that map on-to the Sudoku Puzzle can be represented by a 9-partite graph with 9 vertices with no edges in the same partition which represents 1 colour. Here, the edges between each pair of the 9 partitions can be variable as every vertex inside each partition has 20 choices in making it's connect in the remaining 8 partitions. The edge connect between all the partition pairs, or colourings pair, can be again represented by a $9  9$ matrix which can called a Colouring Matrix $J(C)$.
* As is previously worked out by Sandeep Chowdhary and Collins Assisi, 2019[2] the most stable states against varying excitatory and inhibitory synaptic weights with a constant DC input and without an Oscillatory drive current were the Sudoku Solutions with exactly 2 connections in 4 of the partitions and 3 connections in the rest of 4 partitions, i.e., the $J(C)$ matrix was constituted by 18 and 27 elements alone. In this project our focus is on whether Sudoku solutions will be achieved or not at a given frequency of a Step-function. 
* The other important observations would include if the Sudoku solutions are achieved, what would be the $J(C)$ that will be more stable against the variantion in the input. The time with which these solutions are achieved will also be significant.
* Since, this a first step towards understanding how a more realistic input from the environment which is not a constant, is perceived by the Olfactory circuit still gets recognized as a specific odor, the results will also show us what exactly represent a specific odor, and would also shed light on whether the taken Sudoku-puzzle topology is the best representative model for Olfactory neuron circuits.
